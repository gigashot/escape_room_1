## Příklady metod



* random()  - generuje náhodné reálné číslo od 0 do 1

* uniform(x,y)  - generuje náhodné reálné číslo x<=c<y

* randrange(0,100,2)  - generuje náhodné celé číslo od 0 do 99 sudé

* choice(„abcdefghijk“) - generuje náhodný znak z daného řetězce



```python

#příklady here
mtd_random = random.random()
print(mtd_random)

mtd_uniform = random.uniform(1,5)
print(mtd_uniform)


mtd_choice = random.choice('kočkopes')

mtd_choice
```
